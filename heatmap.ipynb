```python
import pandas as pd

file_path = "/Users/sushmitarajan/Downloads/School Waivers.xlsx"

df = pd.read_excel(file_path, sheet_name="heatmap_all_data")

df["County"] = df["County"].str.strip()  # remove extra spaces
df["county"] = df["County"].str.replace(" County", "", regex=False)

import geopandas as gpd

url = "https://www2.census.gov/geo/tiger/GENZ2023/shp/cb_2023_us_county_500k.zip"

gdf = gpd.read_file(url)

# Filter to Georgia
ga = gdf[gdf["STATEFP"] == "13"]

ga = ga[["NAME", "geometry"]]

merged = ga.merge(df, left_on="NAME", right_on="county", how="left")

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))

merged.plot(
    column="Grand Total",
    cmap="YlOrRd",
    linewidth=0.5,
    edgecolor="black",
    legend=True,
    ax=ax,
    missing_kwds={"color": "lightgrey"}
)

plt.title("All Georgia School Flexibility Waivers (Grand Total)")
plt.axis("off")
plt.show()
```
